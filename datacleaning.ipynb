```python
import pandas as pd

static_df = pd.read_excel("Static.xlsx")
#keep only pacific firms
static_df = static_df[static_df["Region"] == "PAC"]
static_df.head()
```

```python
#import ESCORE ratings
escore_y_df = pd.read_excel("ESCORE_Y.xlsx")
escore_y_df.head()
```

```python
#merge the two dataframes on ISIN inner (to keep only the firms that are in both dataframes aka pacific and have escore)
pac_escore = pd.merge(escore_y_df, static_df, on="ISIN", how="inner")
pac_escore.to_excel("pacific_escore.xlsx", index=False)
pac_escore.head()
```

```python
#drop records colums 1999-2004
dropped = pac_escore.drop(columns=[1999, 2000, 2001, 2002, 2003, 2004, "Name", "Country", "Region"])
dropped
```

```python
#how many firms have 0 nan values (first count the nan values in each row)
dropped["nan_count"] = dropped.isnull().sum(axis=1)
print("There are", dropped[dropped["nan_count"] == 0].shape[0], "firms with complete ESCORE")
```
